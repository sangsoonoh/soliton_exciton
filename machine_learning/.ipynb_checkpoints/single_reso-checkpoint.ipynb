## Time domain analysis

#### Setup the system parameters

```python
params = Params(Nx=1, Ny=1, 
                omega0=1., kappa=1.,
                gamma0=0.001,
                gamma_c=1., delta=1.85, Ns=30.)
print(params)

solver = Solver(params.dt)
print(solver)

source = Source(dict_source={'A': 1, 'omega': params.omega0+params.delta})
print(source)
```

```python
Nsite = params.Nx*params.Ny
H_kerr = HamiltonianKerr(params, HamiltonianSquare)
# H_out = state.get_H_out(2*Nsite, params.gamma_c, list_pos_out=[])
H_out = state.get_H_out(2*Nsite, params.gamma_c, list_pos_out=[0, Nsite])
```

```python
def H_dt(x=None,t=None): return H_kerr(x)
def H_out_dt(x=None,t=None): return H_out
def S_in_dt(x=None,t=None): return state.get_S_in(2*Nsite, source, params.gamma_c, t, list_pos_in=[0, Nsite])
```

#### Time evolution

```python
t_max = 300
Nt = int(t_max / params.dt)
tab_t = np.linspace(0, t_max, Nt)

# init (different/random values for onset of instability)
# psi_init = np.zeros(2*Nsite).astype(np.csingle)
psi_init = np.zeros(2*Nsite).astype(complex)
psi_init[0] = 0.11    # initial value for CW
psi_init[Nsite] = 0.1 # initial value for CCW
S_out = np.empty( (Nt, 2) ).astype(complex)

psi_tspan, H_tspan, S_out = state.evolve(solver, H_dt, H_out_dt, S_in_dt, psi_init=psi_init, tab_t=tab_t, gamma_c = params.gamma_c)
# psi_tspan, H_tspan = state.evolve(solver, H_dt, H_out_dt, S_in_dt, psi_init=None, tab_t=tab_t)
```
